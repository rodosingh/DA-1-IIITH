## Perform Data Cleaning...

```python
# Import data
data = pd.read_csv("football_data.csv", sep = ",")
data.head(7)
```

```python
data.columns
```

```python
# Selecting numeric-columns from the given data.
tmp_data = data[['ID', 'Name', 'Overall', 'Potential', 'Value', 'Wage', 'Special', 'Weak Foot',
       'Skill Moves', 'Position', 'LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW',
       'LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM',
       'CDM', 'RDM', 'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB', 'Crossing',
       'Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys', 'Dribbling',
       'Curve', 'FKAccuracy', 'LongPassing', 'BallControl', 'Acceleration',
       'SprintSpeed', 'Agility', 'Reactions', 'Balance', 'ShotPower', 
       'Jumping', 'Stamina', 'Strength', 'LongShots', 'Aggression',
       'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure',
       'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling',
       'GKKicking', 'GKPositioning', 'GKReflexes']]

tmp_data.head()
```

```python
pd.options.mode.chained_assignment = None
def str_to_float(x): 
    x['Value'] = x['Value'][1:-1] # Defining this function would be longer than just applying "lambda" directly.

# Time to clean some of the columns
# Convert "Value" and "Wage" to numeric.
tmp_data['Value'] = tmp_data['Value'].apply(lambda x:np.float32(x[1:-1]) if len(x)>2 else 0.)
tmp_data['Wage'] = tmp_data['Wage'].apply(lambda x:np.float32(x[1:-1]) if len(x)>2 else 0.)
```

```python
# NOw filter out GoalKeepers from the data and make a different column called "performance" according to the position of player.
tmp_data = tmp_data.assign(GK=lambda x: (x['GKDiving'] + x['GKHandling'] + x['GKKicking'] + x['GKPositioning'] + x['GKReflexes'])/5.)
```

```python
tmp_data
```

```python
# Drop those rows where player are not assigned to any position type.
tmp_data = tmp_data[tmp_data['Position'].notnull()]
```

```python
# Change position ratings to numericals tmp_data['Performance'] = 
tmp_data['Performance'] = tmp_data.apply(lambda x: np.int32(x[x['Position']][:-2]) if x['Position'] != 'GK' else x[x['Position']], axis = 1)
```

```python
# Prepare data for CASE-I
data_case1 = tmp_data[['ID', 'Name', 'Overall', 'Potential', 'Weak Foot',
       'Skill Moves', 'Performance', 'Crossing','Finishing', 'HeadingAccuracy', 'ShortPassing',
       'Volleys', 'Dribbling', 'Curve', 'FKAccuracy', 'LongPassing', 'BallControl', 'Acceleration',
       'SprintSpeed', 'Agility', 'Reactions', 'Balance', 'ShotPower',
       'Jumping', 'Stamina', 'Strength', 'LongShots', 'Aggression',
       'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure',
       'Marking', 'StandingTackle', 'SlidingTackle']]
data_case2 = tmp_data[['ID', 'Name', 'Overall', 'Potential', 'Value', 'Wage', 'Special', 'Weak Foot',
       'Skill Moves', 'Performance', 'Crossing','Finishing', 'HeadingAccuracy', 'ShortPassing',
        'Volleys', 'Dribbling', 'Curve', 'FKAccuracy', 'LongPassing', 'BallControl', 'Acceleration',
       'SprintSpeed', 'Agility', 'Reactions', 'Balance', 'ShotPower',
       'Jumping', 'Stamina', 'Strength', 'LongShots', 'Aggression',
       'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure',
       'Marking', 'StandingTackle', 'SlidingTackle']]
# Prepare data for CASE-II
```

```python
# Normalizing the data, with each column being in the range 0-100
def normalizer(x):
    return np.round(100*(x - min(x))/(max(x) - min(x)), 3)
```

```python
# Transform both data_case1 and data_case2
data_case1_normalized = data_case1.iloc[:, 2:].apply(normalizer)# by-default , axis = 0
data_case2_normalized = data_case2.iloc[:, 2:].apply(normalizer)# by-default , axis = 0
data_case1_normalized['ID'] = data_case1.ID
data_case1_normalized['Name'] = data_case1.Name
data_case2_normalized['ID'] = data_case2.ID
data_case2_normalized['Name'] = data_case2.Name
```
