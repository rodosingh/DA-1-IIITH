## Part 2,4

```python
data = data.dropna(subset=["Club"])
```

```python
clubs = data.groupby('Club')
```

```python
clubs.first()
```

**Detecting Outliers using distribution of players in different clubs by using attribute "Wage"**

```python
arr = [25,50,75]
```

```python
def boxplot_team(team_name):
    team = clubs.get_group(team_name).values
    salary = team[:,12].astype(str)
    
    for i in range(len(salary)):
        salary[i] = salary[i][1:-1]
    
    salary1 = [int(x) for x in salary]
    p_tile = np.percentile(salary1,arr)
    
    Q_tile1 = p_tile[0]
    Q_tile2 = p_tile[1]
    Q_tile3 = p_tile[2]
    
    IQR_ = Q_tile3 - Q_tile1
    mn = Q_tile1 - 1.5*IQR_
    mx = Q_tile3 + 1.5*IQR_
    
    out_index = np.where((salary1 > mx ) | (salary1 < mn))   
    out = team[[out_index],:]
    print("Outliers for " + team_name + " are :")
    if out.size!=0:
        print(out[0][0][0][2])
    else:
        print("None")
    
    return salary1
    
```

```python
psg_wage = boxplot_team("Paris Saint-Germain")
```

```python
mu_wage = boxplot_team("Manchester United")
```

```python
chelsea_wage = boxplot_team("Chelsea")
```

```python
rm_wage = boxplot_team("Real Madrid")
mc_wage = boxplot_team("Manchester City")
fcb_wage = boxplot_team("FC Barcelona")
juv_wage = boxplot_team("Juventus")
lvp_wage = boxplot_team("Liverpool")
```

```python
wages_list = [fcb_wage,juv_wage,mu_wage,mc_wage,rm_wage,chelsea_wage,psg_wage,lvp_wage]
plt.boxplot(wages_list)
plt.xticks([1,2,3,4,5,6,7,8],['FCB','JUV','MU','MC','RM','Chelsea','PSG',"Liverpool"])
plt.title("Boxplot for Clubs based on Wages of Players")
plt.ylabel("Wages in thousand Euros")
```

```python
pos = data.groupby('Position')
```

```python
pos.first()
```

```python
def boxplot_perf(pos_name):
    pos_perf = pos.get_group(pos_name)
    pos_perf_ = pos_perf.loc[:,pos_name].values
    
    ppf = [float(x[:-2]) for x in pos_perf_ ]
    ppf = np.array(ppf)
    
    
    p_tile = np.percentile(ppf,arr)
    
    Q_tile1 = p_tile[0]
    Q_tile2 = p_tile[1]
    Q_tile3 = p_tile[2]
    
    IQR_ = Q_tile3 - Q_tile1
    mn = Q_tile1 - 1.5*IQR_
    mx = Q_tile3 + 1.5*IQR_
    
    
    out_index = np.where((ppf > mx ) | (ppf < mn)) 
    temp = np.array(pos_perf)
    
    cnt = temp[out_index].shape[0]
    
    print("Outliers for position "+pos_name+" are")
    
    if cnt!=0:
        for i in range(cnt):
            print(temp[out_index][i][2])
    else:
        print("None")
    
    return ppf
    
```

```python
diff_pos = ["CF","CB","CM","RF","ST","LAM"]
```

```python
Z=[]
```

```python
for x in diff_pos:
    Z.append(boxplot_perf(x))
    print()
```

```python
ppf_list = [Z[0],Z[1],Z[2],Z[3],Z[4],Z[5]]
plt.boxplot(ppf_list)
plt.xticks([1,2,3,4,5,6],diff_pos)
plt.title("Boxplot for different positions based on players performance in that position")
plt.ylabel("Efficiency (out of 100)")
```
